````python
# Set Measurement Timestamp as index, sort the index, and make sure Measurement Index is in datetime format

df = df.set_index('Measurement Timestamp').sort_index()
print(df.index.duplicated().sum())
df.index = pd.to_datetime(df.index)
df.head()
````

````python
# 2. save csv's
````

````python
# Extract hour, day of wweek, month, year, day name, is weekend from Measurement Timestamp column

df['hour'] = df.index.hour
df['day_of_week'] = df.index.dayofweek
df['month'] = df.index.month
df['year'] = df.index.year
df['day_name'] = df.index.day_name()
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)

df.reset_index().to_csv('output/q3_wrangled_data.csv', index=False)

# turn new temporal features into a csv 

new_df = df.reset_index()[['Measurement Timestamp', 'hour', 'day_of_week', 'month', 'year', 'day_name', 'is_weekend']]

print(new_df.isna().sum())

new_df.to_csv('output/q3_temporal_features.csv', index=False)
````

````python
# output/q3_datetime_info.txt

# save start date and end date of data into a report

earliest_date = df.index.min()
latest_date = df.index.max()

with open('output/q3_datetime_info.txt', 'w') as f:
    f.write(f"Start date: {earliest_date}\n")
    f.write(f"End date: {latest_date}\n")
````
